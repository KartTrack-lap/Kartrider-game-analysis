```python
import numpy as np
import pandas as pd
import requests
import json
import urllib.request
import warnings
from tqdm import tqdm
from datetime import datetime
import time
from datetime import timedelta
warnings.filterwarnings(action='ignore')
```

```python
# 각자의 api key를 복사하여 넣어주세요
api='입력'
```

# 두가지 궁금한점이 있습니다.
- drop_duplicates를 썼을 때와 안쓰고 nunique() 값을 쓸때 왜 값이 다르게 나오는지
- 현재는 2분 기준으로 api를 날리게끔 되어 있는데 1분 단위로 날리면 매치 건수가 다르게 나올지

## 중복제거 버전
- match_df.drop_duplicates(subset='matches', inplace=True) 사용

```python
def get_matchdata_a (start, total_end):
    
    start_date=start
    end_date=datetime.strptime(start_date, '%Y.%m.%d %H:%M:%S')+timedelta(minutes=2)
    end_date = datetime.strftime(end_date, '%Y.%m.%d %H:%M:%S')
    limit='200'
    match_df=pd.DataFrame()
    count=0

    # 마지막 start_date가 total_end-2분까지 돌아가도록 설정
    while start_date != total_end :
        
        # 10,000건에 한번씩 5초 슬립
        if count%50==0:
            time.sleep(5)
            
        # offset 0, 200 둘다 돌리도록
        for j in [0, 200]:
            offset=j
            headers = {'Authorization': api}
            _API_URL = f'https://api.nexon.co.kr/kart/v1.0/matches/all?start_date={start_date}&end_date={end_date}&offset={offset}&limit={limit}'
            res = requests.get(_API_URL, headers = headers)
            data=res.json()

            for i in tqdm(range(0, len(data['matches']))):
                tmp=pd.DataFrame(data['matches'][i]['matches'], columns=['matches'])
                tmp['matchtype']=data['matches'][i]['matchType']
                tmp['start']=start_date
                tmp['end']=end_date
                match_df=pd.concat([match_df, tmp])
            
            count=count+1

        # offeset 0, 200 한번씩 돌아간 뒤에 2분 뒤로 다시 돌릴 수 있도록 start와 end 에 2분씩 더해줌
        new_start=datetime.strptime(start_date, '%Y.%m.%d %H:%M:%S')+timedelta(minutes=2)
        start_date = datetime.strftime(new_start, '%Y.%m.%d %H:%M:%S')
        new_end=datetime.strptime(end_date, '%Y.%m.%d %H:%M:%S')+timedelta(minutes=2)
        end_date = datetime.strftime(new_end, '%Y.%m.%d %H:%M:%S')
    
    # 중복 제거
    match_df.drop_duplicates(subset='matches', inplace=True)
    
    #reset_index 인덱스 초기화
    match_df.reset_index(inplace=True, drop=True)
    
    # api 돌아간 횟수 print(출력)   
    #print(count)     
    return match_df
```

```python
a=get_matchdata_a('2021.06.19 00:00:00', '2021.06.20 00:00:00')
```
